```python
import pandas as pd
import numpy as np
import matplotlib as mp
import glob
```

```python
names = glob.glob('*.csv')

```

```python
combine = pd.DataFrame()
for file in names:
    data = pd.read_csv(file, sep=';', encoding='windows-1251')
    combine = pd.concat([combine,data])
```

```python
combine.shape

```

```python
combine.describe()
```

```python
combine.rename(columns = {'Название':'name','Категория':'category','Бренд':'brend','Остаток':'remainder', 'Отзывов':'Reviews', 'Минимальная цена':' min_price'
                          , 'Максимальная цена':'max_price', 'Средняя цена':'mean_price', 'Продаж':'sales'
                          , 'Выручка':' Revenue', 'Рейтинг':'Rating'}, inplace = True )
```
